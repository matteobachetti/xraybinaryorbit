### In this observation, the Fe xxvi emission line energy exhibited remarkable Doppler shifts. In this example, we will utilize the fit_disc_ps function from the xraybinaryorb package to disentangle this intriguing behavior.

```python
plt.errorbar(phase_array,y, yerr =y_err,fmt="k", alpha=0.5)
plt.errorbar(phase_array,y,fmt="k")
plt.xlabel("Orbital phase")
plt.ylabel("Fe xxi energy evolution")
```

```python
# We DIVIDE THE DATA

x1 = np.array(x[0:6])
y1= np.array(y[0:6])
y_data_err1 = np.array([y_err[0][0:6], y_err[1][0:6]])
phase1 = phase_array[0:6]

x2 = np.array(x[6:25])
y2= np.array(y[6:25])
y_data_err2 = np.array([y_err[0][6:25], y_err[1][6:25]])
phase2 = phase_array[6:25]

x3 = np.array(x[25:30])
y3= np.array(y[25:30])
y_data_err3 = np.array([y_err[0][25:30], y_err[1][25:30]])
phase3= phase_array[25:30]

x4 = np.array(x[30:41])
y4 = np.array(y[30:41])
y_data_err4 = np.array([y_err[0][30:41], y_err[1][30:41]])
phase4 = phase_array[30:41]

#In each calculation the boundaries of iphase should be changed to match these values:
min(phase1),min(phase2),min(phase3),min(phase4)
```

```python
result1, phase1, dphase1, predicted_data1, r_squared1 = fit_disc_ps(x1, y1, y_err=y_data_err1, num_iterations=3, maxiter = 1000, swarmsize = 100, units="keV",method_="extended",extended_binsize=0.01)
```

```python
result1.to_csv('result1.csv', index=False)
np.savetxt('phase1.txt',phase1, delimiter=',')
np.savetxt('dphase1.txt',dphase1, delimiter=',')
np.savetxt('predicted_data1.txt',predicted_data1, delimiter=',')
plt.plot(phase1, predicted_data1,"o:")
plt.errorbar(phase1, y1, yerr=y_data_err1)
dphase1
result1
```

```python
result2, phase2,dphase2,  predicted_data2, r_squared2 = fit_disc_ps(x2, y2, y_err=y_data_err2, num_iterations=3,maxiter = 1000, swarmsize = 100, units="keV",method_="extended",extended_binsize=0.01)
```
